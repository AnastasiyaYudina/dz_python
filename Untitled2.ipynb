```python
import pandas as pd
```

```python
import random
lst = ['robot'] * 10
lst += ['human'] * 10
random.shuffle(lst)
data = pd.DataFrame({'whoAmI':lst})
data.head()
```

```python
unique_values=data['whoAmI'].unique() # Получаем уникальные значения из стобцов 'whoAmI'
```

```python
new_data=pd.DataFrame()# Создаем пустой DataFrame
```

```python
for value in unique_values:
      new_data[value]=(data['whoAmI']==value).astype(int) # Для каждого уникального значения создаем новый столбец и заполняем его значениями 0 или 1
```
